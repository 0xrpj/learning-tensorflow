Defining constants

-> It's a simplest category of a tensor.

-> Can't change.

-> Can't be trained.

-> Can have any dimension.

```python
from tensorflow import constant

# 2x3 constant
a = constant(3, shape=[2,3])

print(a.numpy())

print("---------------")

# 2x2 constant
b = constant([1,2,3,4], shape=[2,2])

print(b.numpy())
```

Summing over tensor dimensions

```python
A = tf.ones([2,3,4])

print(A.numpy())
print(tf.reduce_sum(A))
```

```python
print(tf.reduce_sum(A,0)) #Summing 0D
# Sum is still 24.
```

```python
print(tf.reduce_sum(A,1)) #1D
```

```python
print(tf.reduce_sum(A,2)) #2D
```

## Finding the optimum

We might need to find:
Minimum: Lowest value of a loss function.

Maximum: Highest value of objective function.

gradient() is used in such scenarios.

Optimum: Where gradient = 0

Minimum: Change in gradient > 0 

Maximum: Change in gradient < 0

```python
# Define x

x = tf.Variable(-1,0, dtype=tf.float64)
# 0 Dimensional Variable here.
# Also could be written like this:
# x = tf.Variable(-1.0) :p
# Its float to avoid warnings.

# Define y within instance of GradientTape

with tf.GradientTape() as tape:
  tape.watch(x)
  y = tf.multiply(x,x) # y = x^2
  # Computing the rate of change of y with respect to x.

  # Evaluate the gradient of y at x = -1
  g = tape.gradient(y,x) #dy/dx is stored in g
  print(g.numpy())

#Slope/Derivative is -2 at x is -1.
```
